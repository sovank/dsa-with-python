# **Seive of Eratosthenes**

```python
A = 7
isPrime = [False]* 2 + [True] * (A-1)
print(isPrime)

count = 0
ans = []

for num in range(2, A+1):
  if isPrime[num]:
    for j in range(num*num, A+1, num):
        isPrime[j] = False

print(isPrime)

for i in range(2, len(isPrime)):
  if isPrime[i]:
    ans.append(i)

ans

# T.C ~ O(Alog(log(A)))
# S.C ~ O(A)
```

# Smallest prime Factor

```python
A = 9
primeFactor = [0] * (A+1)

for i in range(2, A+1):
  primeFactor[i] = i

print(primeFactor)

for i in range(2, A+1):
  if primeFactor[i] == i:
    for j in range(i*i, A+1, i):
      primeFactor[j] = i

print(primeFactor[A])


# T.C ~ O(Alog(log(A)))
# S.C ~ O(A)
```

# Prime Factorization

```python
def PrimeFactors(A):
  primeFactor = [0] * (A+1)

  for i in range(2, A+1):
    primeFactor[i] = i

  for i in range(2, int(A**0.5) + 1):
    if primeFactor[i] == i:
      for j in range(i*i, A+1, i):
        if primeFactor[j] == j:
          primeFactor[j] = i
  return primeFactor

A = 10
ans = 1
primeFactors = PrimeFactors(A)
print(primeFactors)
while A>1: #x>1 also
  x = primeFactors[A]
  count =0
  while A % x == 0:
    count += 1
    A = A // x
  ans *= (count + 1)
print(ans)
```
