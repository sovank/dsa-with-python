Need to Solve this using **Seive of Eratosthenes**

```python
A = 7
isPrime = [False]* 2 + [True] * (A-1)
print(isPrime)

count = 0
ans = []

for num in range(2, A+1):
  if isPrime[num]:
    for j in range(num*num, A+1, num):
        isPrime[j] = False

print(isPrime)

for i in range(2, len(isPrime)):
  if isPrime[i]:
    ans.append(i)

ans

# T.C ~ O(Alog(log(A)))
# S.C ~ O(A)
```

Refined code

```python
A = 7
isPrime = [False] * 2 + [True] * (A - 1)
print("Initial isPrime list:", isPrime)

for num in range(2, A + 1):
    if isPrime[num]:
        for j in range(num * num, A + 1, num):
            isPrime[j] = False

print("Final isPrime list:", isPrime)
```
