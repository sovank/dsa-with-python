# Brute Force

```python
def has_zero_sum_subarray(arr):
    n = len(arr)

    # Iterate over all possible starting points of the subarray
    for start in range(n):
        # Iterate over all possible ending points of the subarray
        for end in range(start, n):
            # Calculate the sum of the subarray arr[start:end+1]
            subarray_sum = 0
            for k in range(start, end + 1):
                subarray_sum += arr[k]
            # Check if the sum is zero
            if subarray_sum == 0:
                return True

    return False

# Example usage
arr = [1, 2, -3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: True

arr = [1, 2, 3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: False

# T.C ~ O(N^3)
# S.C ~ O(1)
```

# Using Prefix Sum

```python
def has_zero_sum_subarray(arr):
    n = len(arr)

    # Compute the prefix sum array
    prefix_sum = [0] * n
    prefix_sum[0] = arr[0]
    for i in range(1, n):
        prefix_sum[i] = prefix_sum[i - 1] + arr[i]

    # Iterate over all possible starting points of the subarray
    for start in range(n):
        # Iterate over all possible ending points of the subarray
        for end in range(start, n):
            # Calculate the sum of the subarray arr[start:end+1] using prefix sums
            if start == 0:
                subarray_sum = prefix_sum[end]
            else:
                subarray_sum = prefix_sum[end] - prefix_sum[start - 1]

            # Check if the sum is zero
            if subarray_sum == 0:
                return True

    return False

# Example usage
arr = [1, 2, -3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: True

arr = [1, 2, 3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: False

# T.C ~ O(N^2)
# S.C ~ O(N)
```

# Further Optimizing

```python
def has_zero_sum_subarray(A):
  pfsum = [0] * len(A)
  pfsum[0] = A[0]
  set_Pfsum = set()

  for i in range(1, len(A)):
      pfsum[i] = pfsum[i-1] + A[i]

  for i in range(len(pfsum)):
      if pfsum[i] == 0:
          return 1

  set_Pfsum = set(pfsum)
  if len(set_Pfsum) != len(pfsum):
      return 1

  return 0


# Example usage
arr = [1, 2, -3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: True

arr = [1, 2, 3, 4, 5]
print(has_zero_sum_subarray(arr))  # Output: False

# T.C ~ O(N)
# S.C ~ O(N)
```

```python
def FirstRepeatingElement(A):
  dic_A = {}
  set_A = set(A)

  if len(set_A) == len(A):
      return -1

  for i in range(len(A)):
      if A[i] not in dic_A:
          dic_A[A[i]] = 1
      else:
          dic_A[A[i]] += 1

  for i in range(len(A)):
      if dic_A[A[i]] > 1:
        return(A[i])


A = [10, 5, 3, 4, 3, 5, 6]
#A = [6, 10, 5, 4, 9, 120]

FirstRepeatingElement(A)
```
