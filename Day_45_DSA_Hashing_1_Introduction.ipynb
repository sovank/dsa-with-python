```python
A = [2, 6, 3, 8, 2, 3, 8, 10, 6, 8]
Q = [2, 8, 3, 5]
dic = {}

for num in A:
  if num not in dic:
    dic[num] = 1
  else:
    dic[num] += 1

for num in Q:
  if num in dic:
    print(num, ":", dic[num])
  else:
    print(num, ":", 0)
```
