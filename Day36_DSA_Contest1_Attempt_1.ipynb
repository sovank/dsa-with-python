```python
A = [
    [1, 2],
    [3, 3]
    ]
B = 3
n = len(A)
m = len(A[0])

i = 0
j = m-1

val = 1000000000
while i<n and j>=0:
    if A[i][j] == B:
        val = min(val, (i+1)*1009 + (j+1))
        j-=1
    elif A[i][j] < B:
        i+=1
    else:
        j-=1
if val != 1000000000:
    print(val)
else:
    print(-1)
```

```python
#A = [3, 3, 5, 0, 1]
A = [4, 8, 7, 3, 9, 4, 1]
n = len(A)
maxsum = 0
currsum = A[0]
for i in range(n-1):
    if A[i] > A[i+1]:
        currsum += A[i+1]
        maxsum = max(maxsum, currsum)
    else:
        currsum = A[i+1]
        maxsum = max(maxsum, currsum)
print(maxsum)
```

```python
A = [1, 2, 3]
B = [0, 1, 2]

ans = []
for i in range(len(B)):
    count = 0
    pair = 0
    for j in range(len(A)):
        if A[j] & (1<<B[i]):
            count += 1
    pair = (count *(count-1))//2
    ans.append(pair)
print(ans)
```
