```python
A = [1, 2, 3, 4, 5]
#A = [1, 8, 4, 12, 5, 10, 6]
B = 2

n = len(A)
freq = [0] * B

for i in range(n):
    freq[A[i]%B] += 1 #this will calculate the freq array in shortcut

ans = (freq[0] * (freq[0]-1))//2  # because they(A%B==0) to avoid repetitions (here 2,4 and 4,2 we have to take 1)

i = 1
j = B-1

while i<=j:
  if i == j:
    ans += (freq[i] * (freq[j]-1))//2 # if same index then take one pair without repeating
  else:
    ans += (freq[i] * freq[j]) # else multiply the values
  i += 1
  j -= 1

print(ans)
```

Given an array of integers, return the entire gcd of an array

```python
A = [20, 30, 40]
n = len(A)
arr = []

def gcd(A, B):
  if B==0:
    return A
  return gcd(B, A%B)

def main(A):
  result = A[0]
  for num in A[1:]:
    result = gcd(result, num)

  return result


print(main(A))
```
