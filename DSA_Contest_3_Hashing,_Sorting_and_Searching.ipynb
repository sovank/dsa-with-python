# Painter's problem

image.png

```python
class Solution:
    def solve(self, A, B):
        low = max(A)
        high = sum(A)

        while low < high:
            mid = (low + high) // 2
            if canPack(A, B, mid):
                # it means it's possible to partition the weights with the current mid capacity.
                # Therefore, we adjust our search to look for a potentially smaller capacity, so high = mid.
                high = mid
            else:
                #  it means the current mid capacity is too small to pack the weights into at most B groups
                # Therefore, we need to increase our capacity (low = mid + 1) to continue searching for a feasible solution
                low = mid + 1
        return low

def canPack(weights, B, capacity):
    current_weight = 0
    num_backpacks = 1

    for weight in weights:
        if current_weight + weight > capacity:
            num_backpacks += 1
            current_weight = weight

            if num_backpacks > B:
                return False
        else:
            current_weight += weight

    return True


A = [1,2,3] # weights
B = 2 # num of backpacks

A = [3,2,2,1] # weights
B = 4 # num of backpacks

solution = Solution()
solution.solve(A, B)
```

# Pair Sum

image.png

```python
class Solution:
    def solve(self, A, B):
        uniqueSet = set()
        for num in B:
            if A-num in uniqueSet:
                return 1

            uniqueSet.add(num)

        return 0

A = 8
B = [3,5,1,2,1,2]
solution = Solution()
solution.solve(A, B)
```
