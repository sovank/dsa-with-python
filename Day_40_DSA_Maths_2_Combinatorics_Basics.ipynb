```python
n = 5
pattern = []
for i in range(n):
  row = []
  for j in range(n):
    if i == j or j == 0:
      row.append(1)
    else:
      row.append(0)
  pattern.append(row)

for i in range(2, len(pattern)):
  for j in range(1, i):
    pattern[i][j] = pattern[i-1][j-1] + pattern[i-1][j]

for i in range(len(pattern)):
  print(pattern[i])
```
