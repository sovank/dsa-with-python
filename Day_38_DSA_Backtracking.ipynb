```python
def generate_subsets(arr, current_subset, index, all_subsets):

    # Base condition
    if index == len(arr):
        all_subsets.append(current_subset.copy())
        return

    # Pick
    current_subset.append(arr[index])
    generate_subsets(arr, current_subset, index + 1, all_subsets)

    # No-Pick ---> Exclude the current element (backtrack)
    current_subset.pop()
    generate_subsets(arr, current_subset, index + 1, all_subsets)

# Initialize variables
input_list = [1, 2, 3]
result = []

# Generate subsets
generate_subsets(input_list, [], 0, result)

# Print the result
print(result)
```
